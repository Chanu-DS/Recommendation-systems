#  using weighted average for each movie avg rating

###  w= (rv+cm)/(v+m)
### v- votes r-avg rating c- total rating m-min votes required

```python
v=movies_cleaned_df['vote_count']
r=movies_cleaned_df['vote_average']
c=movies_cleaned_df['vote_average'].mean()
m=movies_cleaned_df['vote_count'].quantile(0.70)
```

```python
movies_cleaned_df['weighted_avg']=((r*v)+(c*m))/(v+m)
```

```python
movies_cleaned_df.head()
```

```python
movies_sorted_ratings_df=movies_cleaned_df.sort_values('weighted_avg',ascending=False)
```

```python
movies_sorted_ratings_df[['original_title','popularity','weighted_avg','overview','revenue']]#.reset_index(drop=True, inplace=True)
```

# now wedo 50% on popularity and 50% on weighted avg

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaling=MinMaxScaler()

movies_scaled=scaling.fit_transform(movies_cleaned_df[['weighted_avg','popularity']])
movies_normalized=pd.DataFrame(movies_scaled,columns=['weighted_avg','popularity'])
movies_normalized.head(30)
```

```python
movies_cleaned_df[['normalized_weighted_avg','normalized_popularity']]=movies_normalized
movies_cleaned_df.head()
```

```python
movies_cleaned_df['score']=movies_cleaned_df['normalized_weighted_avg']*0.5+movies_cleaned_df['normalized_popularity']*0.5
final_recommendation=movies_cleaned_df[['original_title','normalized_weighted_avg','normalized_popularity','score']].sort_values('score',ascending=False)
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x=final_recommendation['score'].head(10),y=final_recommendation['original_title'].head(10),data=final_recommendation)
plt.title('Movie recommendation weighted avg--popularity')
plt.xlabel('score',weight='bold')
plt.ylabel('Movies')
plt.show()
plt.savefig('Votes_avg and popularity top movies list')
```
